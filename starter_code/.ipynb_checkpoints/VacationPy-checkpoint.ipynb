### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
csv_path="../output_data/cities_weathers.csv"
weather_csv_df=pd.read_csv(csv_path)
weather_csv_df["Max Temp"]=round((weather_csv_df["Max Temp"]-273.15)*(9/5)+32,1)
weather_csv_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#plot heatmap
fig = gmaps.figure(center=(26.0, 10.0),zoom_level=1.75)

#sets the location
locations=weather_csv_df[['Lat','Lng']].astype(float)

#sets humidity as weight
rating=weather_csv_df['Humidity'].astype(float)

#generates heatmap
heat_layer=gmaps.heatmap_layer(locations,weights=rating,dissipating=False, max_intensity=200,
                              point_radius=3)
fig.add_layer(heat_layer)

#outputs map
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.

```python
#finds cities with ideal weather and stores in df
ideal_weather=weather_csv_df.loc[(weather_csv_df['Wind Speed']<10) &
                                (weather_csv_df['Cloudiness']==0) &
                                (weather_csv_df['Max Temp']>=70) &
                                (weather_csv_df['Max Temp']<=80)]
ideal_weather
                                
```
